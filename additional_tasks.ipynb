# Задача 1
# Используйте данный URL"https://collectionapi.metmuseum.org/public/collection/v1/search?q=QUERY" где "QUERY" ключевое слово
## Предположим, ваша задача – найти все произведения, связанные с подсолнухами. Выполните поиск по слову "sunflowers". Внимательно посмотрите на формат ответа и сохраните id полученных объектов в список sunflower_ids.

## В качестве ответа выберите id объектов (3), которые присутствуют в полученном списке: [436524,16822570, 20149, 437329, 2032]

```python
# URL для поиска данных в API Метрополитен-музея
url = 'https://collectionapi.metmuseum.org/public/collection/v1/search?q='
IDs = 'sunflowers' # Ключевое слово для поиска ("sunflowers" - подсолнухи)
r = requests.get(url+IDs) # Выполняем GET-запрос на API, добавив ключевое слово к URL
res = r.json() # 'res' теперь содержит результаты поиска в формате JSON, которые можно анализировать и использовать далее
res
```

```python
sunflower_ids = [] # Создаем пустой список для хранения найденных id объектов
# Извлекаем все id объектов из результата поиска по запросу "sunflowers" и сохраняем в список sunflower_ids
# Результат запроса API содержит ключ 'objectIDs', который хранит список id найденных объектов
sunflower_ids = res['objectIDs'] 
```

```python
search_ids = [437329, 20149, 436524, 2032, 16822570] # Список id, которые нужно проверить на наличие в полученном списке sunflower_ids
for id in search_ids: # Проходим по каждому id в search_ids
    if id in sunflower_ids:     # Если текущий id найден в sunflower_ids, выводим сообщение, что элемент найден
        print(f"Элемент {id} найден")
    else:     # Если id не найден, выводим сообщение, что элемент не найден
        print(f"Элемент {id} не найден")
```

# Задача 3 
## Теперь напишите цикл, с помощью которого будет собрана информация об объектах, которые были получены на 2 шаге. Результаты запишите в датафрейм, а именно – следующие параметры:

### objectID – id объекта
### title – название
### artistDisplayName – автор
### department – департамент
### objectBeginDate – дата (начало)
### objectEndDate – дата (конец)
### period – название периода
### objectName – название/категория объекта
### culture – культура
### Сколько объектов относятся к японской культуре? (culture –Japan)

```python
# Пример запроса на получение списка объектов, связанных с подсолнухами
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?q=sunflowers')
res = r.json()
```

```python
# Сохраняем список id объектов
sunflower_ids = res.get('objectIDs', [])
```

```python
# Создаем пустой датафрейм для итоговых данных
full_df = pd.DataFrame()
```

```python
# Цикл по всем объектам из списка sunflower_ids
for i in sunflower_ids:
    # Выполняем запрос по каждому объекту
    r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{i}')
    res = r_test.json()
    
    # Извлекаем информацию об объекте, используя метод .get() для безопасного доступа к значениям
    df = pd.DataFrame({
        'objectID': [res.get('objectID')],
        'title': [res.get('title')],
        'artistDisplayName': [res.get('artistDisplayName')],
        'department': [res.get('department')],
        'objectBeginDate': [res.get('objectBeginDate')],
        'objectEndDate': [res.get('objectEndDate')],
        'period': [res.get('period')],
        'objectName': [res.get('objectName')],
        'culture': [res.get('culture')]
    })
    
    # Добавляем данные в итоговый датафрейм
    full_df = pd.concat([full_df, df], ignore_index=True)
```

```python
# Печатаем полученные данные
full_df
```

```python
full_df.query('culture == "Japan"') \
    .agg({'objectID': 'count'})
```

# Ответ: к Японской культуре относится 1 объект 

# Задача 4
## Выполните поисковый запрос для департамента "Asian art" (departmentId = 6) с тегом "cat" и запишите id объектов в cat_ids. Используйте цикл из шага 4, чтобы получить данные о первой 1000 объектов из списка.
### В качестве ответа укажите, в каком периоде (period) было выполнено наибольшее количество работ, представленных в музее?

```python
# Отправляем запрос на API Метрополитен-музея с поисковым запросом "cat" в отделе с departmentId=6 (например, отдел "Египетское искусство")
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=6&q=cat')
cats = r.json() # Преобразуем ответ API в формат JSON
cats_id = [] # Создаем пустой список для хранения ID объектов, которые соответствуют запросу
cats_id = cats['objectIDs'] # Получаем список всех ID объектов, найденных по запросу, из ключа 'objectIDs'
```

```python
# Ограничиваем список первыми 1000 объектами (чтобы избежать работы с очень большим количеством данных)
cats_id = cats_id[:1000]
```

```python
# Создаем пустой датафрейм для итоговых данных
full_df_cats_id = pd.DataFrame()
```

```python
# Цикл по всем объектам из списка cats_id
for i in cats_id:
    # Выполняем запрос по каждому объекту
    r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{i}')
    res = r_test.json()
    
    # Извлекаем информацию об объекте, используя метод .get() для безопасного доступа к значениям
    df = pd.DataFrame({
        'objectID': [res.get('objectID')],
        'title': [res.get('title')],
        'artistDisplayName': [res.get('artistDisplayName')],
        'department': [res.get('department')],
        'objectBeginDate': [res.get('objectBeginDate')],
        'objectEndDate': [res.get('objectEndDate')],
        'period': [res.get('period')],
        'objectName': [res.get('objectName')],
        'culture': [res.get('culture')]
    })
    
    # Объединяем текущий df с итоговым датафреймом
    full_df_cats_id = pd.concat([full_df_cats_id, df], ignore_index=True)
```

```python
# Печатаем полученные данные
full_df_cats_id
```

```python
# Посчитаем в каком периоде (period) было выполнено наибольшее количество работ, представленных в музее.
full_df_cats_id.period.value_counts()
```
